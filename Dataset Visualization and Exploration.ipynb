```python
# PROBLEM STATEMENT
# The objective of this task is to explore and visualize the Iris dataset to
# understand the distribution of features, relationships between variables,
# and identify potential outliers.

# Data manipulation
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)

print("Libraries imported successfully!")
```

```python
# Load Iris dataset from seaborn

iris = sns.load_dataset('iris')

print("Dataset loaded successfully!")
```

```python
print("Dataset Shape:")
print(iris.shape)
```

```python
print("\nColumn Names:")
print(iris.columns.tolist())
```

```python
print("\nFirst Five Rows:")
iris.head()
```

```python
print("Dataset Information:")
iris.info()
```

```python
print("Summary Statistics:")
iris.describe()
```

```python
print("Missing Values:")
print(iris.isnull().sum())
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=iris,
    x='sepal_length',
    y='petal_length',
    hue='species',
    s=100
)

plt.title('Sepal Length vs Petal Length')
plt.show()
```

```python
sns.pairplot(
    iris,
    hue='species'
)

plt.show()
```

```python
iris.hist(
    figsize=(12,8),
    bins=15
)

plt.suptitle("Feature Distributions")
plt.show()
```

```python
plt.figure(figsize=(12,6))

for i, column in enumerate(iris.columns[:-1]):

    plt.subplot(2,2,i+1)

    sns.boxplot(
        y=iris[column]
    )

    plt.title(f'Boxplot of {column}')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    iris.drop('species', axis=1).corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')
plt.show()
```

```python
print("""
Final Insights:

1. The Iris dataset contains 150 observations and 5 columns.
2. No missing values were found.
3. Petal length and petal width show strong positive correlation.
4. Setosa species is clearly separable from the other species.
5. Histograms show feature distributions are mostly well-behaved.
6. Boxplots reveal a few possible outliers in sepal measurements.
7. Petal measurements appear to be the most useful features for classification.
""")
```
